```python
import pandas as pd
#skjdhfsjds dfgdfsg
data = pd.read_csv("train.csv")
pd.set_option('display.max_rows', 500)
data.rename(columns = {'1stFlrSF':'_1stFlrSF'}, inplace = True)
data.rename(columns = {'2ndFlrSF':'_2ndFlrSF'}, inplace = True)
data.rename(columns = {'3SsnPorch':'_3SsnPorch'}, inplace = True)
data['MSZoning'].unique()

import pandas as pd
import sqlite3
import os

# if os.path.exists('test_database.db'):
#     os.remove('test_database.db')
    

conn = sqlite3.connect('test_database.db')
c = conn.cursor()

c.execute("PRAGMA ignore_check_constraints=0")
c.execute("PRAGMA foreign_keys = 1")

# c.execute('DROP TABLE train')
# c.execute('DROP TABLE MSSubClass_Type')
# c.execute('DROP TABLE MSZoning_Type')


c.execute('CREATE TABLE IF NOT EXISTS MSSubClass_Type(Type int PRIMARY KEY NOT NULL UNIQUE)')
c.execute('''INSERT INTO MSSubClass_Type(Type) VALUES (20),(30),(40),(45),(50),(60),
(70),(75),(80),(85),(90),(120),(150),(160),(180),(190)''')


c.execute('CREATE TABLE  IF NOT EXISTS MSZoning_Type(Type TEXT PRIMARY KEY NOT NULL UNIQUE)')
c.execute('''INSERT INTO MSZoning_Type(Type) VALUES ('A'), ('C'), ('FV'), ('I'), ('RH'), ('RK'), ('RP'), ('RM')''')


c.execute('CREATE TABLE  IF NOT EXISTS Street_Type(Type TEXT PRIMARY KEY NOT NULL UNIQUE)')
c.execute('''INSERT INTO Street_Type(Type) VALUES ('Grvl'), ('Pave')''')

c.execute('CREATE TABLE  IF NOT EXISTS Alley_Type(Type TEXT PRIMARY KEY NOT NULL UNIQUE)')
c.execute('''INSERT INTO Alley_Type(Type) VALUES ('Grvl'), ('Pave'), ('NA')''')

c.execute('CREATE TABLE  IF NOT EXISTS LotShape_Type(Type TEXT PRIMARY KEY NOT NULL UNIQUE)')
c.execute('''INSERT INTO LotShape_Type(Type) VALUES ('Reg'), ('IR1'), ('IR2'), ('IR3')''')

c.execute('CREATE TABLE  IF NOT EXISTS LandContour_Type(Type TEXT PRIMARY KEY NOT NULL UNIQUE)')
c.execute('''INSERT INTO LandContour_Type(Type) VALUES ('Lvl'), ('Bnk'), ('HLS'), ('Low')''')

c.execute('CREATE TABLE  IF NOT EXISTS Utilities_Type(Type TEXT PRIMARY KEY NOT NULL UNIQUE)')
c.execute('''INSERT INTO Utilities_Type(Type) VALUES ('AllPub'), ('NoSewr'), ('NoSeWa'), ('ELO')''')

c.execute('CREATE TABLE  IF NOT EXISTS LotConfig_Type(Type TEXT PRIMARY KEY NOT NULL UNIQUE)')
c.execute('''INSERT INTO LotConfig_Type(Type) VALUES ('Inside'), ('Corner'), ('CulDSac'), ('FR2'), ('FR3')''')

c.execute('CREATE TABLE  IF NOT EXISTS LandSlope_Type(Type TEXT PRIMARY KEY NOT NULL UNIQUE)')
c.execute('''INSERT INTO LandSlope_Type(Type) VALUES ('Gtl'), ('Mod'), ('Sev')''')

c.execute('CREATE TABLE  IF NOT EXISTS Condition_Type(Type TEXT PRIMARY KEY NOT NULL UNIQUE)')
c.execute('''INSERT INTO Condition_Type(Type) VALUES ('Artery'), ('Feedr'), ('Norm'), ('RRNn'),
('RRAn'), ('PosN'), ('PosA'), ('RRNe'), ('RRAe')''')

c.execute('CREATE TABLE  IF NOT EXISTS BldgType_Type(Type TEXT PRIMARY KEY NOT NULL UNIQUE)')
c.execute('''INSERT INTO BldgType_Type(Type) VALUES ('1Fam'), ('2FmCon'), ('Duplx'), ('TwnhsE'), ('TwnhsI')''')

c.execute('CREATE TABLE  IF NOT EXISTS HouseStyle_Type(Type TEXT PRIMARY KEY NOT NULL UNIQUE)')
c.execute('''INSERT INTO HouseStyle_Type(Type) VALUES ('1Story'), ('1.5Fin'), ('1.5Unf'), ('2Story'), ('2.5Fin'),
('2.5Unf'), ('SFoyer'), ('SLvl')''')

c.execute('CREATE TABLE  IF NOT EXISTS RoofStyle_Type(Type TEXT PRIMARY KEY NOT NULL UNIQUE)')
c.execute('''INSERT INTO RoofStyle_Type(Type) VALUES ('Flat'), ('Gable'), ('Gambrel'), ('Hip'), ('Mansard'),
('Shed')''')

c.execute('CREATE TABLE  IF NOT EXISTS RoofMatl_Type(Type TEXT PRIMARY KEY NOT NULL UNIQUE)')
c.execute('''INSERT INTO RoofMatl_Type(Type) VALUES ('ClyTile'), ('CompShg'), ('Membran'), ('Metal'), ('Roll'),
('Tar&Grv'), ('WdShake'), ('WdShngl')''')

c.execute('CREATE TABLE  IF NOT EXISTS Exterior_Type(Type TEXT PRIMARY KEY NOT NULL UNIQUE)')
c.execute('''INSERT INTO Exterior_Type(Type) VALUES ('AsbShng'), ('AsphShn'), ('BrkComm'), ('BrkFace'), ('CBlock'),
('CemntBd'), ('HdBoard'), ('ImStucc'), ('MetalSd'), ('Other'), ('Plywood'), ('PreCast'), ('Stone'),
('Stucco'), ('VinylSd'), ('Wd Sdng'), ('WdShing')''')


c.execute('CREATE TABLE  IF NOT EXISTS MasVnrType_Type(Type TEXT PRIMARY KEY NOT NULL UNIQUE)')
c.execute('''INSERT INTO MasVnrType_Type(Type) VALUES ('BrkCmn'), ('BrkFace'), ('CBlock'), ('None'), ('Stone')''')


c.execute('''CREATE TABLE  IF NOT EXISTS train (
Id INT PRIMARY KEY, 
MSSubClass INT CHECK( MSSubClass IN (20,30,40,45,50,60,70,75,80,85,90,120,150,160,180,19) ), 
MSZoning text NOT NULL, 
LotFrontage FLOAT, LotArea INT, 
Street text, 
Alley text, 
LotShape text, 
LandContour text, 
Utilities text, 
LotConfig text, 
LandSlope text, 
Neighborhood text, 
Condition1 text, 
Condition2 text, 
BldgType text, HouseStyle text, 
OverallQual INT CHECK(OverallQual>=1 AND OverallQual<=10), 
OverallCond INT CHECK(OverallCond>=1 AND OverallCond<=10), 
YearBuilt INT, YearRemodAdd INT, 
RoofStyle text, 
RoofMatl text, Exterior1st text, 
Exterior2nd text, MasVnrType text, MasVnrArea FLOAT, ExterQual text, ExterCond text, 
Foundation text, BsmtQual text, BsmtCond text, BsmtExposure text, BsmtFinType1 text, 
BsmtFinSF1 INT, BsmtFinType2 text, BsmtFinSF2 INT, BsmtUnfSF INT, TotalBsmtSF INT, 
Heating  text, HeatingQC text, CentralAir text, Electrical text, _1stFlrSF INT, 
_2ndFlrSF INT, LowQualFinSF INT, GrLivArea INT, BsmtFullBath INT, BsmtHalfBath INT, 
FullBath INT, HalfBath INT, BedroomAbvGr INT, KitchenAbvGr INT, KitchenQual text, 
TotRmsAbvGrd INT, Functional text, Fireplaces INT, FireplaceQu text, GarageType text, 
GarageYrBlt FLOAT, GarageFinish text, GarageCars INT, GarageArea INT, GarageQual text, 
GarageCond text, PavedDrive text, WoodDeckSF INT, OpenPorchSF INT, EnclosedPorch INT, 
_3SsnPorch INT, ScreenPorch INT, PoolArea INT, PoolQC text, Fence text, MiscFeature text, 
MiscVal INT, MoSold INT, YrSold INT, SaleType text, SaleCondition text, SalePrice INT,
FOREIGN KEY (MSZoning) REFERENCES MSZoning_Type(Type))''')

conn.commit()

data.to_sql('train', conn, if_exists='replace', index = False)
 
c.execute('''  
SELECT * FROM train
          ''')

for row in c.fetchall():
    print (row)
```
